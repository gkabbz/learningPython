### Filtering DataFrames By One Condition

To see a sample of the data

```python
titanic.head(10)
```

To see a sample of the data for a specific column

```python
titanic.sex.head(10)
```

To see a series based off a condition on a row

```python
titanic.sex == "male"
```

To apply the series condition to a dataframe and create a new dataframe

```python
titanic[titanic.sex=="male"]
```

Same condition can be applied using the .loc method. 
The advantage of using the .loc method is you can also filter for a specific column at the same time. Not using the loc method means you'd have to do chained indexing like this: `titanic[titanic.sex=="male"]["fare"]`
Not great because it makes the code harder to read

```python
titanic.loc[titanic.sex == "male", "fare"]
```

For better readability, you can assign the condition (series) to a variable. Then apply the series to a dataframe

```python
titanic_male_series = titanic.sex == "male"
titanic_male_passengers = titanic.loc[titanic_male_series]
titanic_male_passengers.head(10)
```

```python
titanic.dtypes
```

```python
string_column_types = titanic.dtypes == object
string_column_types
```

```python
titanic.loc[:, ~string_column_types]
```

```python
titanic.loc[titanic_male_series, string_column_types]
```

### Filtering DataFrames with Many Conditions (AND)

```python
# Get a series for the condition you are interested in
male_passengers = titanic.sex == "male"
```

```python
# Series for next condition
passengers_older_than_14 = titanic.age > 14
```

```python
# You can combine series together using AND
(male_passengers & passengers_older_than_14).head()
```

```python
# You can then use the combined series conditions to apply to a dataframe to select a new dataframe
# that has the conditions you want to examine
male_survivors = titanic.loc[male_passengers & passengers_older_than_14, ['survived', 'pclass', 'sex', 'age']]
```

```python
male_survivors
```

```python
male_survivors.describe()
```

### Filtering DataFrames by Many Conditions (OR)

```python
# Check if passenger is female
female_passengers = titanic.sex == "female"
# Check if passenger is child
passengers_less_than_14_years = titanic.age <= 14
```

```python
women_or_children = titanic.loc[female_passengers | passengers_less_than_14_years, 
                               ["survived", "pclass", "sex", "age"]]
```

```python
women_or_children.head()
```

```python
women_or_children.describe()
```

```python
titanic.describe()
```

### Advanced Filtering with between(), isin() and ~

```python
summer.head()
```

```python
olympics_1988_condition = summer.Year == 1988
olympics_1988 = summer.loc[olympics_1988_condition]
```

```python
olympics_1988.head()
```

```python
olympics_1988.tail()
```

```python
olympic_games_after_1992_condition = summer.Year >= 1992
olympic_games_after_1992 = summer.loc[olympic_games_after_1992_condition]
```

```python
olympic_games_after_1992.head()
```

```python
olympic_games_after_1992.tail()
```

```python
condition_between_60_and_69 = summer.Year.between(1960, 1969, inclusive=True)
olympic_games_60s = summer.loc[condition_between_60_and_69]
```

```python
olympic_games_60s.head()
```

```python
olympic_games_60s.tail()
```

```python
favourite_years = [1972, 1996]
condition_favourite_years = summer.Year.isin(favourite_years)
favourite_olympic_years = summer.loc[condition_favourite_years]
```

```python
favourite_olympic_years.head()
```

```python
favourite_olympic_years.tail()
```

```python
not_favourite_olympic_years = summer.loc[~condition_favourite_years]
```

```python
not_favourite_olympic_years.head()
```

```python
not_favourite_olympic_years.Year.unique()
```

### Checking for values using any() and all()

```python
titanic.sex == "male"
```

```python
# Check if any of the passengers were male
# Will return true if condition is found
(titanic.sex == "male").any()
```

```python
# Check if all the passengers were male
# Will return False if all passengers are not male
(titanic.sex == "male").all()
```

```python
# Check to see if there was a passenger who was exactly 80 years old
(titanic.age == 80).any()
```

```python
condition_80_years_old = titanic.age == 80
```

```python
#Check to see if there were any 80 year old passengers
condition_80_years_old.any()
```

```python
passengers_80_yrs_old = titanic.loc[condition_80_years_old]
passengers_80_yrs_old
```
